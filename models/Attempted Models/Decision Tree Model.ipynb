```python
data_1 = data.copy()
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
data_1['Legendary_le'] = lb.fit_transform(data_1['Legendary'])
```

```python
data_1.tail(10)
```

```python
# One hot encoding
data_1 = pd.get_dummies(data_1,columns=['Type 1','Type 2'],prefix=['Type_1','Type_2'])
```

```python
data_1.head()
```

```python
# Drop Column
data_1.drop(['#','Name'],axis=1, inplace=True)
```

```python
data_1.drop('Legendary',axis=1, inplace=True)
```

```python
data_1.info()
```

```python
# Split train test data 
from sklearn.model_selection import train_test_split
```

```python
x = data_1.loc[:,data_1.columns != 'Legendary_le']
y = data_1.loc[:,data_1.columns == 'Legendary_le']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
#Visualiizing Skew
count_classes = pd.value_counts(data_1['Legendary_le'],sort=True).sort_index()
count_classes
```

```python
count_classes.plot(kind='bar')
```

```python
# Decision Tree FINALLLLLLLLLLLLY 
# *crossing fingers* 
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
# Evaluation time
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)
```
